```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import openpyxl
import japanize_matplotlib # 追加


iterations = list(range(1, 101))  # 1～100回のイテレーション
scores = [100 - i + (i % 10) for i in iterations]  # 評価値の例（変動するスコア）

# グラフを描画
plt.figure(figsize=(10, 6))
plt.plot(iterations, scores, marker='o', linestyle='-', color='blue', label='評価値')

# 軸ラベルとタイトル
plt.xlabel('iterations')
plt.ylabel('score')
plt.title('イテレーションごとの評価値の推移')
plt.grid(True)
plt.legend()
plt.tight_layout()

# 表示
plt.show()
```

```python

# データ生成
iterations = list(range(1, 101))  # 1～100回のイテレーション
times = [i * 0.2 for i in range(1, 101)]  # 仮の時間（0.2秒ごとに増加）

# グラフを描画（横軸：時間、縦軸：イテレーション回数）
plt.figure(figsize=(10, 6))
plt.plot(times, iterations, marker='o', linestyle='-', color='green', label='イテレーション数')

# 軸ラベルとタイトル
plt.xlabel('時間 (秒)')
plt.ylabel('イテレーション回数')
plt.title('時間ごとのイテレーション数の推移')
plt.grid(True)
plt.legend()
plt.tight_layout()

# 表示
plt.show()
```
