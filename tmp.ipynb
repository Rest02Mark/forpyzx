```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib


# --- データの定義 ---
# スレッド数
threads = np.array([1, 2, 4, 8, 16])

# 各スレッド数での平均実行時間 (秒)
times = np.array([0.840310, 0.421148, 0.217244, 0.117375, 0.062017])


# --- スピードアップ率の計算 ---
# 基準となる1スレッドでの実行時間
baseline_time = times[0] 
# スピードアップ率 = (1スレッドでの実行時間) / (Nスレッドでの実行時間)
speedup_rates = baseline_time / times


# --- グラフの描画 ---
# 日本語フォントの設定（お使いの環境に合わせてフォント名を変更してください）
# Windows: 'Yu Gothic', 'Meiryo'など
# Mac: 'Hiragino Sans'など
# Linux: 'IPAexGothic'など (要インストール: sudo apt-get install fonts-ipaexfont)
plt.rcParams['font.family'] = 'IPAexGothic'

# グラフのサイズを指定
plt.figure(figsize=(8, 6))

# 1. 実際のスピードアップ率をプロット (青い線、マーカー付き)
plt.plot(threads, speedup_rates, marker='o', linestyle='-', color='blue', label='実測スピードアップ率')

# 2. 理想的なスピードアップ率をプロット (赤い破線)
# 理想的なスピードアップ率はスレッド数と同じ (例: 4スレッドなら4倍)
plt.plot(threads, threads, linestyle='--', color='red', label='理想スピードアップ率 (N)')

# --- グラフの装飾 ---
# タイトル
plt.title('スレッド数とスピードアップ率', fontsize=16)

# 軸ラベル
plt.xlabel('スレッド数', fontsize=12)
plt.ylabel('スピードアップ率 (倍)', fontsize=12)

# X軸の目盛りをデータに合わせて設定
plt.xticks(threads)

# グリッド線を表示
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 凡例を表示
plt.legend(fontsize=12)

# グラフを画面に表示
plt.show()

# --- 参考：計算結果の表示 ---
print("--- 計算結果 ---")
for i in range(len(threads)):
    print(f"スレッド数: {threads[i]}, スピードアップ率: {speedup_rates[i]:.4f}倍")


plt.savefig("/home/rest/forpyzx/speedup.pdf", format="pdf")
```
